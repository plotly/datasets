```python
import plotly.offline as py
import plotly.graph_objs as go
import pandas as pd
import math

py.init_notebook_mode()
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv")
```

```python
df_2007 = data[data['year']==2007]
```

```python
df_2007 = df_2007.sort(['continent', 'country'])
```

```python
slope = 2.666051223553066e-05
```

```python
hover_text = []
bubble_size = []

for index, row in df_2007.iterrows():
    hover_text.append('Country: {country}<br> Life Expectancy: {lifeExp}<br>GDP per capita: {gdp}'
               '<br>Population: {pop}<br>Year: {year}'.format(country=row['country'], 
                                                              lifeExp=row['lifeExp'], 
                                                              gdp=row['gdpPercap'], 
                                                              pop=row['pop'],
                                                              year=row['year']))
    bubble_size.append(math.sqrt(row['pop']*slope))
    
```

```python
df_2007['text'] = hover_text
df_2007['size'] = bubble_size
```
